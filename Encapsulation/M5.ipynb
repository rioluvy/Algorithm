```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
## 특정 행, 열 제거하기
# drop() 함수 내에 삭제하려는 column이름이나 index를 작성
# axis = 1(열을 제거하는 경우), axis = 0(행을 제거하는 경우), 기본값은 1

print('Before : ', iris.columns.tolist())
iris = iris.drop(['sepal_width'],axis=1)
print('After : ', iris.columns.tolist())
```

```python
## 특정 행, 열의 합 구하기
# sum() 함수를 활용해 특정 행,열의 총 합을 계산한다.
# 특정 열에 대한 정보만 얻거나 여러 열의 합을 구할 수도 있다.
# axis = 0(행), axis = 1(열) 계산할 축 선택, 기본값은 1
iris.sum()
```

```python
## 매우 중요
## 중복치 처리: duplicated()
## 데이터 내에 존재하는 중복된 행 or 열 확인
## sum()함수와 함께 사용해 중복된 갯수를 얻는다
iris['species'].duplicated().head()
```

```python
## drop_duplicates()
## drop_duplicates(column_name_list): 중복된 행,열을 제거한다.
# 기준으로 하는 열들의 이름 리스트
# keep: 중복된 것들 중 남길 항목 선정['first','second','False]
iris.drop_duplicates(['petal_length'],keep = 'first').head()
```

```python
## indexing
# loc, iloc
# index를 사용하여 직접 행이나 열을 선택할 수 있다.
# loc: 조건문이나 label을 통해 행,열을 선택
# iloc: index를 기반으로 행, 열을 선택
iris.loc[10,'species']

iris.iloc[10,3]

iris.iloc[:,3].head()
```

```python
## 문자열 처리
# 공백 제거하기: str.strip(), lstrip(), rstrip()
# 문자열 앞, 뒤에 공백이 들어가 있는 경우 제거
# strip: 앞, 뒤 모두 공백 제거
# l & r + strip(): 앞 또는 뒤 의 공백만 제거
## df['species'].str.strip()
```

```python
## 구분자로 문자열 나누기: str.split(구분자)
# 예를 들어 abcd@gmail.com 을 @로 나누는 경우 abcd와 gmail.com으로 두개의 컬럼 생성
# split()후 strip()을 사용하여 에러 방지 가능
#iris.petal_width_str.str.split('.')
```

```python
## 패턴찾기
# str.startswith(pat): 문자열이 pat으로 시작하는지 여부를 반환
# str.endswith(pat): 문자열이 pat으로 끝나는지 여부를 반환
# str.contains(pat): 문자열이 pat을 포함하는지 여부를 반환
# str.match(pat:regex): 정규표현식인 pat을 만족하는지 여부를 반환
iris[iris.species.str.contains('ini')].head()
```

```python
## 그룹지어 연산하기: groupby() + agg()
# groupby: 이후 여러 연산, 통계를 한번에 구하고 싶은 경우 agg() 함수를 한번더 활용한다
# agg()의 인자로 리스트가 오는 경우: 모든 컬럼에 대해 해당 연산을 모두 진행
# agg()의 인자로 dict가 오는 경우: 해당하는 컬럼마다 dict의 연산만 진행
iris.groupby('species').agg(['mean','var'])
```

```python
import pandas as pd

def analyze_class_members2(data):
    # 1. Dict 형태로 주어진 수강생 데이터를 pandas dataframe으로 변경
    df = pd.DataFrame(data)

    # 2. 성별이 '남성'이고 나이가 35 이상인 데이터 추출
    filtered_df = df[(df['성별'] == '남성') & (df['나이'] >= 35)]['나이'].reset_index(drop = True)

    # 3. 성별을 기준으로 평균 나이 계산
    mean_age_by_gender = df.groupby('성별')['나이'].mean()

    # 3. 성별을 기준으로 최고 나이 추출
    max_age_by_gender = df.groupby('성별')['나이'].max()

    return filtered_df, mean_age_by_gender, max_age_by_gender

# 테스트
data = {'이름': ['Alice', 'Bob', 'Charlie', 'David'],
        '나이': [25.0, 30.0, 35.0, 40.0],
        '성별': ['여성', '남성', '남성', '남성']}

result = analyze_class_members2(data)

print("성별이 '남성'이고 나이가 35 이상인 수강생:\n", result[0])
print("성별을 기준으로 계산한 평균 나이:\n", result[1])
print("성별을 기준으로 한 최고 나이:\n", result[2])
```

```python
import pandas as pd

# 데이터
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Ella'],
        'Age': [25, 30, 35, 40, 45],
        'Gender': ['F', 'M', 'M', 'M', 'F']}

# Pandas DataFrame 생성
df = pd.DataFrame(data)

# 'Gender' 컬럼을 기준으로 그룹화하여 'Age' 컬럼의 최대값과 최소값 계산

grouped_df = df.groupby('Gender')['Age'].agg(['max','min'])
# 결과를 하나의 DataFrame으로 합치기


# 인덱 재정렬
result_df = grouped_df.stack()

result_df.index.names = ['Gender','Age']

# 출력
print(result_df)
```
